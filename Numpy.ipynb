## Reshape

the reshape  is used to change the size of the matrix or array which is already created 

```python
# In this example i just made the 2-D array as 2,3 first and then change into the (3,2)

a155=np.array([[1,2,3],[4,5,6]])
print(a155.shape)
print(a155)
a155=a155.reshape(3,2)
print(a155.shape)
print(a155)
```

# Example for the 3-D array 

###
 For the 3-D array we have to use one thing to make  sure we can make the product of (slice , row , column ) are also eqaul to the product of after reshape (slice , row , column)

```python

# in this state i just created the one 3 - D array in the shape of (2,2,3)
a166=np.array([[[1,2,3],[4,5,6]],[[6,5,4],[3,2,1]]])
print(a166.shape)
print(a166)


# this one is the reshape part and if you can see that i just reshape the (2,2,3) into the (1,2,6)
a166=a166.reshape(1,2,6)
print(a166.shape)
print(a166)
```

## ndmin


### the ndmin is used to change the uni-dimension array into the mulit-dimension array

```python
# For an example we are going to chnage the uni-dimension array into the mulit - dimension  array 
import numpy as np 
a111=np.array([1,2,3],ndmin=2)
print(a111.ndim) # this one is used to print the what dimension is that array 7
print(a111) #this one is used to print the converted array from the uni-dimension into mulit - dimension

# so  for now we are going to make one uni-dimension array into the multi-dimension array 
# so we change the one dimension array into the 64 dimension array 
a121=np.array([1,2,3,4],ndmin=64)
print(a121.ndim)
print(a121)
print(a121.shape)
```

## arange

### the arange is used to generate  the numbers between the specific limits like who we used in the py programming 

```python
# For  example we are gonna to print the number in the range of 1 to 10 

a1211=np.arange(1,10)
print(a1211)

# so if we are gonna to use the one value in side the ( ) mean it will consider as an end_value 

a1212=np.arange(10)
print(a1212)

# so if we are gonna use  the three values inside the ( ) mean it will consider  the third value as step_value 

a1213=np.arange(0,10,3)
print(a1213)

# like in the range if we use some non print able  range mean it will consider as an empty array 

print(" use the non printable range ")

a1214=np.arange(10,1,1)
print(a1214)

# now we are gonna make that 2-D or 3-D array while using the  arange( ) function 

# this is for 2-D while using the arange( )

print("The 2-D array ")

a1215=np.arange(0,10).reshape(2,5)
print(a1215)

# this is for make the 3-D function while using rh arange ( )

print("The 3-D array ")

a1216=np.arange(0,10).reshape(1,2,5)
print(a1216)
```

## linspace()

### the linspace( ) is used to Generate equally spaced  values between the specific range

```python
# for an example we are gonna to generate the values from 1 to 10 then just print the 5  values  

a1313=np.linspace(1,10,5)
print(a1313)

# now we are gonna to reshape  this into the 2 D array

a1314=np.linspace(1,10,5).reshape(1,5)
print(a1314)
```
