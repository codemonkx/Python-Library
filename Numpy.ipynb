```python
import numpy as np 
np.__version__
```

```python
l=[1,2,3]
print(type(l))
print(l)
# now we are convert the list into the array
a=np.array(l)
print(type(a))
print(a)
```

```python
a3=np.array([True,False])
print(a3.dtype) #bool
print(a3)

a4=np.array([True,False,10])
print(a4.dtype) # int
print(a4)

a5=np.array([True,10,20.5])
print(a5.dtype) # float
print(a5)

a6=np.array([True,10,10+3j])
print(a6.dtype) # Complext
print(a6)

a7=np.array([10,20,30,"Hi"])
print(a7.dtype) # str
print(a7)

a8=np.array([10,20,{"hello bahi",10,30}])
print(a8.dtype) # object
print(a8)

a9=np.array([10,20,{"age":20,"Name":22}])
print(a9.dtype) #object
print(a9)
```
